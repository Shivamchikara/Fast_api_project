```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='latin1')
df.shape
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
# GENDER
```

```python
M = sns.countplot(x='Gender', data=df,palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sels_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
M=sns.barplot(x='Gender', y='Amount', data=sels_gen, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
#AGE GROUP
```

```python
sns.countplot(x='Age Group', data=df, palette="Set2",hue='Gender')
```

```python
M=sns.countplot(x='Age Group', data=df, palette="Set2",hue='Gender')
for bars in M.containers:
  M.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sale_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
M = sns.barplot(x='Age Group', y='Amount', data=sale_age, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
#STATE
```

```python
df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)
sns.set(rc={'figure.figsize':(24,5)})
M = sns.barplot(x='State', y='Amount', data=sales_state, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
#orders
```

```python
sales_orders=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(24,5)})
M = sns.barplot(x='State', y='Orders', data=sales_orders, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
#Marital STATUS
```

```python
sales_mrt_status=sns.countplot(x='Marital_Status', data=df, palette="Set2")
for bars in sales_mrt_status.containers:
  sales_mrt_status.bar_label(bars)
```

```python
sales_status=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(24,5)})
M = sns.barplot(x='Marital_Status', y='Amount', data=sales_status, palette="Set2", hue='Gender')
for bars in M.containers:
  M.bar_label(bars)
```

```python
#Occupations
```

```python
sns.set(rc={'figure.figsize':(20,5)})
M = sns.countplot(x='Occupation', data=df, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
salse_data=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
M = sns.barplot(x='Occupation', y='Amount', data=salse_data, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
# Product Category
```

```python
sns.set(rc={'figure.figsize':(25,5)})
M = sns.countplot(x='Product_Category', data=df, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
sales_data=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
M = sns.barplot(x='Product_Category', y='Amount', data=sales_data, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)
```

```python
sales_data=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
M = sns.barplot(x='Product_ID', y='Orders', data=sales_data, palette="Set2")
for bars in M.containers:
  M.bar_label(bars)

```
